```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso

# Create sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Define alpha values
alpha_values = [0, 10, 20, 40, 100, 10000]
```

```python
# Ridge Regression visualization
plt.figure(figsize=(12, 6))

for alpha in alpha_values:
    ridge_reg = Ridge(alpha=alpha)
    ridge_reg.fit(X, y)
    plt.plot(X, ridge_reg.predict(X), label=f'λ = {alpha}')

plt.scatter(X, y, color='red', label='Data points')
plt.title('Ridge Regression')
plt.xlabel('Weight')
plt.ylabel('Size')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Lasso Regression visualization
plt.figure(figsize=(12, 6))

for alpha in alpha_values:
    lasso_reg = Lasso(alpha=alpha, max_iter=10000)
    lasso_reg.fit(X, y)
    plt.plot(X, lasso_reg.predict(X), label=f'λ = {alpha}')

plt.scatter(X, y, color='red', label='Data points')
plt.title('Lasso Regression')
plt.xlabel('Weight')
plt.ylabel('Size')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cost function visualization for Ridge and Lasso Regression
slope_values = np.linspace(-1, 1, 100)
lambda_values = [0, 10, 20, 40, 400]
```

```python
# Ridge
plt.figure(figsize=(12, 6))

for alpha in lambda_values:
    cost_ridge = [(slope - 0.5) ** 2 + alpha * slope ** 2 for slope in slope_values]
    plt.plot(slope_values, cost_ridge, label=f'λ = {alpha}')

plt.title('Sum of Squared Residuals + λ * Slope² (Ridge)')
plt.xlabel('Slope Values')
plt.ylabel('Cost')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Lasso
plt.figure(figsize=(12, 6))

for alpha in lambda_values:
    cost_lasso = [(slope - 0.5) ** 2 + alpha * abs(slope) for slope in slope_values]
    plt.plot(slope_values, cost_lasso, label=f'λ = {alpha}')

plt.title('Sum of Squared Residuals + λ * |Slope| (Lasso)')
plt.xlabel('Slope Values')
plt.ylabel('Cost')
plt.legend()
plt.grid(True)
plt.show()
```
